```python
#create the first mating pool
def Create_Intial_Population():
    
    first_population=[]
    for counter in range(0,1000):
        random_integer=random.randint(0, 100)
        solution={}
        solution['elbow']=[]
        solution['arm']=[]
        solution['knee']=[]
        solution['wrist']=[]
        if random_integer>0:
            
            #make counter within range 0 - 11
            index=random.randint(0, 11)
            
            #get excersises
            elbow_ex=elbow_list[index]
            knee_ex=knee_list[index]
            upper_arm_ex=upper_arm_list[index]
            
            #encode the data (from strings to bit)
            age=''
            type_of_injury=''
            if elbow_ex[2]=='Adult':
                age='0'
            elif elbow_ex[2]=='Child':
                age='1'
            if elbow_ex[1]=='Stroke':
                type_of_injury='1000'
            elif elbow_ex[1]=='Brain injuries':
                type_of_injury='1001'
            elif elbow_ex[1]=='Spinal cord injuries':
                type_of_injury='1111'
            solution['elbow'].append([type_of_injury,age,elbow_ex[0]])
            ###############################################
            age=''
            type_of_injury=''
            if knee_ex[2]=='Adult':
                age='0'
            elif knee_ex[2]=='Child':
                age='1'
            if knee_ex[1]=='Stroke':
                type_of_injury='1000'
            elif knee_ex[1]=='Brain injuries':
                type_of_injury='1001'
            elif knee_ex[1]=='Spinal cord injuries':
                type_of_injury='1111'
            solution['knee'].append([type_of_injury,age,knee_ex[0]])
            ###############################################
            age=''
            type_of_injury=''
            if upper_arm_ex[2]=='Adult':
                age='0'
            elif upper_arm_ex[2]=='Child':
                age='1'
            if upper_arm_ex[1]=='Stroke':
                type_of_injury='1000'
            elif upper_arm_ex[1]=='Brain injuries':
                upper_arm_ex='1001'
            elif upper_arm_ex[1]=='Spinal cord injuries':
                type_of_injury='1111'
            #print(upper_arm_ex[0])
            solution['arm'].append([type_of_injury,age,upper_arm_ex[0]])
            ###############################################
            age=''
            type_of_injury=''
            if index < 11:
                wrist_ex=wrist_list[index]
                if wrist_ex[2]=='Adult':
                    age='0'
                elif wrist_ex[2]=='Child':
                    age='1'
                if wrist_ex[1]=='Stroke':
                    type_of_injury='1000'
                elif wrist_ex[1]=='Brain injuries':
                    upper_arm_ex='1001'
                elif wrist_ex[1]=='Spinal cord injuries':
                    type_of_injury='1111'
                solution['wrist'].append([type_of_injury,age,wrist_ex[0]])
            else:
                solution['wrist']=[]
        if random_integer>60:
            index=random.randint(0, 11)
            #print(index)
            #get excersises
            elbow_ex=elbow_list[index]
            knee_ex=knee_list[index]
            upper_arm_ex=upper_arm_list[index]
            
            #encode the data (from strings to bit)
            age=''
            type_of_injury=''
            if elbow_ex[2]=='Adult':
                age='0'
            elif elbow_ex[2]=='Child':
                age='1'
            if elbow_ex[1]=='Stroke':
                type_of_injury='1000'
            elif elbow_ex[1]=='Brain injuries':
                type_of_injury='1001'
            elif elbow_ex[1]=='Spinal cord injuries':
                type_of_injury='1111'
            solution['elbow'].append([type_of_injury,age,elbow_ex[0]])
            ###############################################
            age=''
            type_of_injury=''
            if knee_ex[2]=='Adult':
                age='0'
            elif knee_ex[2]=='Child':
                age='1'
            if knee_ex[1]=='Stroke':
                type_of_injury='1000'
            elif knee_ex[1]=='Brain injuries':
                type_of_injury='1001'
            elif knee_ex[1]=='Spinal cord injuries':
                type_of_injury='1111'
            solution['knee'].append([type_of_injury,age,knee_ex[0]])
            ###############################################
            age=''
            type_of_injury=''
            if upper_arm_ex[2]=='Adult':
                age='0'
            elif upper_arm_ex[2]=='Child':
                age='1'
            if upper_arm_ex[1]=='Stroke':
                type_of_injury='1000'
            elif upper_arm_ex[1]=='Brain injuries':
                upper_arm_ex='1001'
            elif upper_arm_ex[1]=='Spinal cord injuries':
                type_of_injury='1111'
            #print(upper_arm_ex[0])
            solution['arm'].append([type_of_injury,age,upper_arm_ex[0]])
        first_population.append(solution)
    return first_population
```

```python
pooling_mate=Create_Intial_Population()

# '1' for Child
# '1001' for Brain injuries
# 0,1,2,1 number of ex
read_input=['0','1001',0,1,2,1]
#calculate fitness in solution
def Compute_Fitness(input_ , indvidual):
    ex_for_elbow=indvidual['elbow']
    ex_for_knee=indvidual['knee']
    ex_for_arm=indvidual['arm']
    ex_for_wrist=indvidual['wrist']
    #find the number of ex in the individual
    ex_for_elbow_len=len(ex_for_elbow)
    ex_for_knee_len=len(ex_for_knee)
    ex_for_arm_len=len(ex_for_arm)
    ex_for_wrist_len=len(ex_for_wrist)
    age=[]
    type_of_injuries=[]
    
    #wights of the fitness
    age_wieght=0
    number_wight=0
    injury_type_wight=0
    for key,value in indvidual.items():
        for v in value:
            age.append(v[1])
            type_of_injuries.append(v[0])
    if ex_for_elbow_len==0 or ex_for_knee_len==0 or ex_for_arm_len==0:
        return 0
    if input_[0] == age[0]:
        for i in age:
            if i==input_[0]:
                age_wieght=1
    if input_[1] == type_of_injuries[0]:
        for i in type_of_injuries:
            if i==input_[1]:
                injury_type_wight=1
    if input_[2]==ex_for_wrist_len:
        number_wight=number_wight+1
    if input_[3]==ex_for_arm_len:
        number_wight=number_wight+1
    if input_[4]==ex_for_knee_len:
        number_wight=number_wight+1
    if input_[5]==ex_for_elbow_len:
        number_wight=number_wight+1
    number_wight=number_wight/4
    #fitness
    f=(number_wight*0.5+age_wieght*0.5+injury_type_wight)/3
    return f

for individual in pooling_mate:
    
    f=Compute_Fitness(read_input,individual)
    print(individual)
    print()
    print(f)
```
